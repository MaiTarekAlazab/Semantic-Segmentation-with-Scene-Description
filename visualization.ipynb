```python
#compare 
import numpy as np
from tensorflow import keras
import tensorflow as tf
from tensorflow.keras import backend as K
import cv2 
import scipy.misc
import matplotlib.pyplot as plt
import imageio
import glob
import os

def dice_coef(y_true, y_pred):
    smooth = 1
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2.0 * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def dice_coef_loss(y_true, y_pred):
    return 1.0 - dice_coef(y_true, y_pred)

def jacard_coef(y_true, y_pred):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (intersection + 1.0) / (K.sum(y_true_f) + K.sum(y_pred_f) - intersection + 1.0)

def jacard_coef_loss(y_true, y_pred):
    return -jacard_coef(y_true, y_pred)  # -1 ultiplied as we want to minimize this value as loss function

def set_model_type(mtype):
    print("successfully changed model type to", mtype)
    global modeltype
    modeltype = mtype

def load_model():
    if(modeltype == "unet"):
        print("the loaded model is unet")
        dependencies = {
        'dice_coef': dice_coef,
        'jaccard_coef_loss': jacard_coef_loss,
        'jaccard_coef':jacard_coef,
        'dice_coef_loss': dice_coef_loss
        }
        model = tf.keras.models.load_model('unet.h5', custom_objects = dependencies)
        print("unet model successfully loaded")
        return model
    
    elif(modeltype == "pspnet"):
        print("the loaded model is pspnet")
        dependencies = {
            'dice_coef': dice_coef,
            'jaccard_coef_loss': jacard_coef_loss,
            'jaccard_coef':jacard_coef,
            'dice_coef_loss': dice_coef_loss
        }
        model = keras.models.load_model('psp.h5', custom_objects = dependencies)
        print("pspnet model successfully loaded")
        return model
    
    elif(modeltype == "deeplabv3"):
        print("the loaded model is deeplabv3")
        dependencies = {
            'dice_coef': dice_coef,
            'jaccard_coef_loss': jacard_coef_loss,
            'jaccard_coef':jacard_coef,
            'dice_coef_loss': dice_coef_loss
        }
        model = keras.models.load_model('deeplab.h5', custom_objects = dependencies)
        print("deeplabv3 model successfully loaded")
        return model
    
    else:
        print("the loaded model is undefined, choosing unet instead")
        dependencies = {
            'dice_coef': dice_coef,
            'jaccard_coef_loss': jacard_coef_loss,
            'jaccard_coef':jacard_coef,
            'dice_coef_loss': dice_coef_loss
        }
        model = tf.keras.models.load_model('/home/t/iti/gp/image-segmentation/model/pspfinal1.h5', custom_objects = dependencies)
        print("unet model successfully loaded")
        return model
```

```python
input_dir = 'dataset/data_semantics/training/image_2'
target_dir = 'dataset/data_semantics/training/semantic'
```

```python
train_images = []
for img_path in sorted(glob.glob(os.path.join(input_dir, "*.png"))):
    img = cv2.imread(img_path)       
    img = cv2.resize(img, (128, 128))
    cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    train_images.append(img)
```

```python
train_masks = [] 
for mask_path in sorted(glob.glob(os.path.join(target_dir, "*.png"))):
    mask = cv2.imread(mask_path,0)       
    mask = cv2.resize(mask, (128, 128), interpolation = cv2.INTER_NEAREST)
    cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
    train_masks.append(mask)
```

```python
train_images = np.array(train_images , dtype='float32')
train_masks = np.array(train_masks , dtype= 'float32')
```
